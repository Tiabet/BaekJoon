```python
k, n = map(int, input().split())
klist = []

for _ in range(k):
    klist.append(int(input()))

start = 1
last = max(klist)
a = 0

while(start<=last):
    mid = (start + last) // 2
    count = 0
    for i in range(k):
        count+= klist[i] // mid
    if count>=n:
        start = mid+1
    else:
        last = mid-1
    a+=1

print(last)
print(a)
```
