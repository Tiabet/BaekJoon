```python
while True:
    s = input()
    if s == '.':
        break
    stack = []
    for i in s:
        if i == '(' or i == '[':
            stack.append(i)
        elif i == ')':
            if not stack or stack[-1] != '(':
                print('no')
                break
            stack.pop()
        elif i == ']':
            if not stack or stack[-1] != '[':
                print('no')
                break
            stack.pop()
    else:
        if not stack:
            print('yes')
        else:
            print('no')
```
