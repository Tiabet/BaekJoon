

최대 길이의 랜선을 원하는 수만큼 만들어야 하는 문제다. 수를 하나씩 키워가면서 찾아도 되겠지만 시간상 너무 오래 걸리기 때문에 이진 탐색 알고리즘을 사용하면 될 것 같다.

```python
k, n = map(int, input().split())
klist = []

for _ in range(k):
    klist.append(int(input()))

start = 1
last = max(klist)
a = 0

while(start<=last):
    mid = (start + last) // 2
    count = 0
    for i in range(k):
        count+= klist[i] // mid
    if count>=n:
        start = mid+1
    else:
        last = mid-1
    a+=1

print(last)
print(a)
```

이진 탐색을 이용해서 필요한 갯수 n보다 적게 나오면 아랫쪽 범위로, 많이 나오면 윗쪽 범위로 탐색 범위를 바꿔가면서 정답을 찾아나간다.
