산술평균, 중앙값, 최빈값, 범위를 출력하는 문제다. 어렵지 않게 풀 수 있을 것 같다.

```python
from collections import Counter

n = int(input())

lst = []

for _ in range(n):
    num = int(input())
    lst.append(num)


print(round(sum(lst)/n))

lst.sort()
print(lst[n//2])

freq = Counter(lst).most_common()

if len(freq)>1 and freq[0][1] == freq[1][1]:
    print(freq[1][0])
else:
    print(freq[0][0])

print(max(lst)-min(lst))
```
