```python
n = int(input())
people = []
for i in range(n):
    x, y = map(int, input().split())
    people.append((x, y))

result = []
for i in range(n):
    count = 1
    for j in range(n):
        if i != j and people[i][0] < people[j][0] and people[i][1] < people[j][1]:
            count += 1
    result.append(count)

print(*result)
```
