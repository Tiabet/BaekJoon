덩치를 계산하라는 특이한 문젠데, 키와 몸무게 모두 크면 덩치가 크다고 판단한다.

```python
n = int(input())
people = []
for i in range(n):
    x, y = map(int, input().split())
    people.append((x, y))

result = []
for i in range(n):
    count = 1
    for j in range(n):
        if i != j and people[i][0] < people[j][0] and people[i][1] < people[j][1]:
            count += 1
    result.append(count)

print(*result)
```

*result 로 편하게 결과를 출력할 수 있었다.
