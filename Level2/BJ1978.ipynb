이 문제는 주어진 수들 중 소수의 개수를 찾아내는 문제이다. 다소 시간이 오래 걸릴 것 같긴 하지만 2부터 모든 수들을 나눠가면서 체크를 해야 할 것 같다.

```python
n = int(input())
nums = map(int, input().split())

count=0

for i in nums :
    found = True
    if i>1:
        for j in range(2,i):
            if i%j == 0:
                found = False
        if found :
            count+=1

print(count)
```

1보다 클때만 for문으로 나눗셈을 진행하여 소수를 카운트한다.
