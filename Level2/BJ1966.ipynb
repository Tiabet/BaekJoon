```python
import sys
input = sys.stdin.readline
```

import heapq

n = int(input())

for _ in range(n):
    pq = []
    num, tofind = map(int,input().split())
    lst = list(map(int,input().split()))
    
    for i in range(num):
        if tofind==i:
            what = lst[i]
        heapq.heappush(pq,10-lst[i])
    
    for j in range(num):
        if pq[j]==10-what:
            print(j+1)
            break

처음에 문제를 잘못 이해해버렸다. 문서를 인쇄하지 않고 queue의 맨 뒤의 재배치한다는 조건을 못 읽었다. deque를 이용해 처음부터 다시 풀어보겠다.

```python
from collections import deque
import sys

t = int(input())

for i in range(t):
    n, m = map(int, input().split())
    queue = deque(list(map(int, sys.stdin.readline().split())))
    count = 0
    while queue:
        best = max(queue)  
        front = queue.popleft() 
        m -= 1 

        if best == front: 
            count += 1 
            if m < 0: 
                print(count)
                break

        else:  
            queue.append(front) 
            if m < 0 : 
                m = len(queue) - 1 
```

찾고자 하는 index를 한칸씩 앞으로 당겨오면서, max와 맨 앞의 값이 같아지면 카운트하고 값을 출력하는 알고리즘이다.
