import heapq

n = int(input())

for _ in range(n):
    pq = []
    num, tofind = map(int,input().split())
    lst = list(map(int,input().split()))
    
    for i in range(num):
        if tofind==i:
            what = lst[i]
        heapq.heappush(pq,10-lst[i])
    
    for j in range(num):
        if pq[j]==10-what:
            print(j+1)
            break

```python
from collections import deque
import sys

t = int(input())

for i in range(t):
    n, m = map(int, input().split())
    queue = deque(list(map(int, sys.stdin.readline().split())))
    count = 0
    while queue:
        best = max(queue)  
        front = queue.popleft() 
        m -= 1 

        if best == front: 
            count += 1 
            if m < 0: 
                print(count)
                break

        else:  
            queue.append(front) 
            if m < 0 : 
                m = len(queue) - 1 
```
