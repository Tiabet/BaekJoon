스택을 구현하면 되는 문제다. 데이터구조론 시간에 한 적이 있기 때문에 어렵지 않게 구현이 될 것 같다.

```python
import sys

n = int(sys.stdin.readline())

stack = []

for _ in range(n):
    word = sys.stdin.readline().split()
    order = word[0]
    
    if order == 'push':
        stack.append(word[1])
    elif order == 'pop':
        if len(stack)==0:
            print(-1)
        else:
            print(stack.pop())
    elif order == 'size' :
        print(len(stack))
    elif order == 'empty':
        if len(stack)==0:
            print(1)
        else:
            print(0)
    elif order == 'top':
        if len(stack)==0:
            print(-1)
        else:
            print(stack[-1])
```

시간 제한이 걸려 있어서 input이 아닌 sys.stdin.readline을 사용해야했다.
