이분 탐색을 통해 적절하게 잘라내는 높이를 찾아주는 문제다.

```python
n, m = map(int, input().split())

tree = list(map(int, input().split()))

left = 0
right = max(tree)

while left<right:
    mid = (left + right)//2
    length = 0
    for i in tree:
        if i - mid >0:
            length+=i-mid
    
    if length < m:
        right = mid
    else:
        left = mid+1

print(left-1)
```
