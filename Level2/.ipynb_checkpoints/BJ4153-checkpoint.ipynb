세 수를 입력받아 직각삼각형이 되는지 확인하면 되는 간단한 문제다.

```python
while True:
    a, b, c = map(int, input().split())

    list=[a,b,c]
    list.sort()
    
    if a==0:
        break
    
    ortho = ((list[2]**2)==(list[1]**2)+(list[0]**2))
    if ortho :
        print('right')
    else :
        print('wrong')
```
