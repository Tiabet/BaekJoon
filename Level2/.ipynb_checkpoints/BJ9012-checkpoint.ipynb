스택을 이용해 들어온 괄호 문자열이 VPS인지를 판단하는 문제다.

```python
import sys
input = sys.stdin.readline
```

```python
n = int(input())

for _ in range(n):
    stnum = 0
    ps = input()
    ps = list(ps)
    for i in range(len(ps)):
        if ps[i]=='(':
            stnum+=1
            
        else :
            stnum -=1
            
            if stnum<0:
                break
    if stnum==0:
        print("YES")
    else:
        print("NO")
```

스택을 직접 구현하지 않고 개념만 따서 이용해 보았다.
