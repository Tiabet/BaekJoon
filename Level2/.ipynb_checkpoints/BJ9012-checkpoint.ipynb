```python
import sys
input = sys.stdin.readline
```

```python
n = int(input())

for _ in range(n):
    stnum = 0
    ps = input()
    ps = list(ps)
    for i in range(len(ps)):
        if ps[i]=='(':
            stnum+=1
            
        else :
            stnum -=1
            
            if stnum<0:
                break
    if stnum==0:
        print("YES")
    else:
        print("NO")
```
