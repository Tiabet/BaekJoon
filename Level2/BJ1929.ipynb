```python
n, m = map(int, input().split())

def getsosu (x):
    found = True
    for i in range(2,x):
        if x%i == 0 :
            found = False
            break
    return found

result = []

for i in range(n,m+1):
    if getsosu(i):
        result.append(i)
        
for i in result:
    print(i)
```

```python
import math
m,n = map(int,input().split(" "))

result = [True for i in range(n + 1)] 

for i in range(2, int(math.sqrt(n)) + 1): #   2   n  
    if result[i] == True:
        j = 2 
        while i * j <= n:
            result[i * j] = False
            j += 1

for i in range(m, n + 1):
    if result[i]:
        if i != 1:
            print(i)
```
