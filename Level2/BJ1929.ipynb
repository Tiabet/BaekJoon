특정 구간에 있는 소수를 출력하는 문제다. 파이썬으로는 이런 걸 쉽게 구현할 수가 있어서 좋은 것 같다.

```python
n, m = map(int, input().split())

def getsosu (x):
    found = True
    for i in range(2,x):
        if x%i == 0 :
            found = False
            break
    return found

result = []

for i in range(n,m+1):
    if getsosu(i):
        result.append(i)
        
for i in result:
    print(i)
```

시간 초과가 계속 나서 찾아봤더니 다른 방식을 사용해야 했다. 아리스토텔레스의 체라고 불리는데, 소수의 배수들을 지워나가면서 소수를 찾는 알고리즘이었다.

```python
import math
m,n = map(int,input().split(" "))

result = [True for i in range(n + 1)] 

for i in range(2, int(math.sqrt(n)) + 1): #소수가 아니려면 최소 2배수는 가능하므로 범위를 n의 루트로 제한
    if result[i] == True:
        j = 2 
        while i * j <= n:
            result[i * j] = False
            j += 1

for i in range(m, n + 1):
    if result[i]:
        if i != 1:
            print(i)
```
