ACM 호텔
엘리베이터에서 가까운 방 먼저 손님들에게 배정해주는 문제다. 저층부터 올라가면서 배정해주면 되기 때문에, 작은 수부터 호텔의 방을 배정해주는 문제가 되겠다.

```python
t = int(input())

for _ in range(t):
    list=[]
    room=101
    h, w, n = map(int, input().split())
    for i in range(w):
        for j in range(h):
            list.append(room)
            room += 100
        room-=100*(h)
        room+=1
    print(list[n-1])
```

3중 for문이 되어서 좋은 코딩이라고 할 수는 없지만 어렵지 않게 답을 구할 수 있었다.
