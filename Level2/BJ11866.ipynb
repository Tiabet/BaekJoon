  .        pop  append        .

```python
from collections import deque

n, k = map(int, input().split())

dq = deque()
result = []

for i in range(1,n+1):
    dq.append(i)

while dq :
    for _ in range(k-1):
        dq.append(dq.popleft())
    result.append(dq.popleft())

formatted_list = "<" + ", ".join(str(i) for i in result) + ">"

print(formatted_list)
```
