요세푸스 순열이라는 문제다. 사람들이 원을 이룬다는 점을 이용해 큐를 만들어서 pop 하자마자 append를 해서 맨 뒤로 보내버리면 자연스러운 구현이 가능할 것이다.

```python
from collections import deque

n, k = map(int, input().split())

dq = deque()
result = []

for i in range(1,n+1):
    dq.append(i)

while dq :
    for _ in range(k-1):
        dq.append(dq.popleft())
    result.append(dq.popleft())

formatted_list = "<" + ", ".join(str(i) for i in result) + ">"

print(formatted_list)
```
