```python
l = int(input())
st = input()

has = []

for char in st:
    has.append(char)

result = 0
for i in range(l):
    tochange = ord(has[i])-96
    num = tochange*(31**i)
    result+=num

while result>1234567891:
    result%=1234567891

print(result)
```
