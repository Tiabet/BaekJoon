```python
l = int(input())
st = input()

has = []

for char in st:
    has.append(char)

result = 0
for i in range(l):
    tochange = ord(has[i])-96
    num = tochange*(31**i)
    result+=num

while result>1234567891:
    result%=1234567891

print(result)
```

파이썬의 ord 함수를 이용해 알파벳을 아스키코드 넘버로 변환할 수 있었다.
