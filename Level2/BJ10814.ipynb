sort 함수를 이용하는 간단한 문제다.

```python
n = int(input())

arr =[]

for k in range(n):
    age, name = input().split()
    arr.append((int(age),k,name))

arr = sorted(arr, key = lambda x: (x[0],x[1]))

for i in range(len(arr)):
    print(arr[i][0],arr[i][2])
```
