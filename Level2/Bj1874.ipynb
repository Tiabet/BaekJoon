```python
n = int(input())

stack = []
answer = []
find = True
cur= 1

for _ in range(n):
    a = int(input())
    while cur <=a:
        stack.append(cur)
        answer.append('+')
        cur+=1
    
    if stack[-1]==a:
        stack.pop()
        answer.append('-')
    else :
        find = False
        break;

if find :
    for i in answer:
        print(i)
else :
    print("NO")
```
