
스택의 개념을 이용해서 주어진 배열을 만드는 문제이다.

```python
n = int(input())

stack = []
answer = []
find = True
cur= 1

for _ in range(n):
    a = int(input())
    while cur <=a:
        stack.append(cur)
        answer.append('+')
        cur+=1
    
    if stack[-1]==a:
        stack.pop()
        answer.append('-')
    else :
        find = False
        break;

if find :
    for i in answer:
        print(i)
else :
    print("NO")
```

스택을 미리 만들어놓는 것이 아닌, 입력 받을 때마다 수를 스택에 쌓아가는 것이 핵심이다.
원하는 수를 만났을 때까지 while 문을 통해 push를 해주고, 원하는 수를 만나면 pop을 해주는 것이다.
이때 5 - 3 - 4 처럼 감소했다 다시 증가하면 NO를 출력한다. pop을 함으로써 이미 수를 버렸기 때문이다.
