수를 찾는 간단한 문젠데, 메모리의 문제 때문에 여러 작업을 해줘야 할 것 같다.

```python
n = int(input())
arr = list(map(int, input().split()))
m = int(input())
tofind = list(map(int, input().split()))

result = []
arr.sort()

def search_x(nums, x):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == x:
            return 1
        elif nums[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    
    return 0

for i in tofind:
    found = search_x(arr,i)
    result.append(found)

for s in result:
    print(s)
```
