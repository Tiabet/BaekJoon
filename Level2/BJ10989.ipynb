```python
n=int(input())

result=[]

for _ in range(n):
    result.append(int(input()))

result.sort()

for i in range(n):
    print(result[i])
```

```python
import sys

n = int(sys.stdin.readline())
num_list = [0] * 10001

for _ in range(n):
    num_list[int(sys.stdin.readline())] += 1

for i in range(10001):
    if num_list[i] != 0:
        for j in range(num_list[i]):
            print(i)
```
