```python
n = int(input())

meeting = []

for _ in range(n):
    start, end = map(int, input().split())
    meeting.append([start, end])

meeting.sort(key = lambda x : (x[1],x[0])) # .      .

count = 1
meet_end = meeting[0][1]

for i in range(1,n):
    if meeting[i][0]>=meet_end:
        count+=1
        meet_end = meeting[i][1]
        
print(count)
```
