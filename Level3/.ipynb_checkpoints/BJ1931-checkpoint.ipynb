문제는 간단한데 구현하기엔 살짝 까다로운 면이 있는 것 같다. 동적 프로그래밍으로 푸는 건 아닌 것 같아서 찾아보니 그리디 알고리즘이라는 개념을 사용한 것이었다. 좀 더 공부를 해보니 이 문제가 대표적인 케이스였다. 그리디 알고리즘의 핵심은 같은 간격에 최대한 많은 작업을 실행하는 것이다. 따라서 필요한 것은 최대한 빨리 끝나는 작업들을 선택하면 되는 것이다.

```python
n = int(input())

meeting = []

for _ in range(n):
    start, end = map(int, input().split())
    meeting.append([start, end])

meeting.sort(key = lambda x : (x[1],x[0])) #시작시간도 고려해줘야한다. 시작함과 동시에 끝나는 회의도 있기 때문.

count = 1
meet_end = meeting[0][1]

for i in range(1,n):
    if meeting[i][0]>=meet_end:
        count+=1
        meet_end = meeting[i][1]
        
print(count)
```
