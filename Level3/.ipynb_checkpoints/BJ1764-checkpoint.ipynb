이름이 굉장히 재밌는 문제다. 하지만 핵심은 결국 공집합의 개념을 응용하는 것이다. 명단을 사전순으로 출력해야 하므로 sort 함수도 사용해야 할 것이다.

```python
import sys
#input = sys.stdin.readline

listen = []
see = []

n,m = map(int, input().split())

for i in range(n):
    listen.append(input().strip())

for j in range(m):
    see.append(input().strip())
    
intersection = sorted(list(set(listen) & set(see)))

print(len(intersection))
for str in intersection:
    print(str)
```

strip 함수를 사용해서 줄바꿈 문자들을 제거해줄 필요가 있었다. 그리고 set의 & 연산자를 통해 쉽게 교집합을 구할 수 있었다.
