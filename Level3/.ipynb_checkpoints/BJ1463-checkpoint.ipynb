다이나믹 프로그래밍을 이용하는 문제다. 

```python
n = int(input())
d=[0]*(n+1) #편의성을 위해 d[1] 부터 시작하기로 한다.

for i in range(2, n+1):
    d[i] = d[i-1]+1
    if i%2 == 0:
        d[i] = min(d[i],d[i//2]+1)
    if i%3 == 0:
        d[i] = min(d[i],d[i//3]+1)

print(d[n])
```

동적 프로그래밍은 모든 가능성을 저장해두는 프로그래밍이다. 보통 배열에 저장하는데, 이 문제 같은 경우는 앞의 수들의 최단 경로를 저장해놓고 그 뒤의 최단경로들을 빠르게 찾아내는 방식으로 풀 수 있었다.
