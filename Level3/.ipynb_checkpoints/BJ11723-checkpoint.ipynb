```python
import sys

n = int(sys.stdin.readline())
Set = set()

for _ in range(n):
    temp = sys.stdin.readline().strip().split()
    
    if len(temp) == 1:
        if temp[0] == "all":
            Set = set([i for i in range(1, 21)])
        else:
            Set = set()
    
    else:
        order, x = temp[0], temp[1]
        x = int(x)

        if order == "add":
            Set.add(x)
        elif order == "remove":
            Set.discard(x)
        elif order == "check":
            print(1 if x in Set else 0)
        elif order == "toggle":
            if x in Set:
                Set.discard(x)
            else:
                Set.add(x)
```
