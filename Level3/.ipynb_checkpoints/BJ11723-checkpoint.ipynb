이미 있으면 연산을 무시한다는 것을 보아 그냥 리스트가 아닌 Set을 사용해야 함을 알 수 있다. 나머진 평범하므로 빠르게 풀 수 있을 것이다.

```python
import sys

n = int(sys.stdin.readline())
Set = set()

for _ in range(n):
    temp = sys.stdin.readline().strip().split()
    
    if len(temp) == 1:
        if temp[0] == "all":
            Set = set([i for i in range(1, 21)])
        else:
            Set = set()
    
    else:
        order, x = temp[0], temp[1]
        x = int(x)

        if order == "add":
            Set.add(x)
        elif order == "remove":
            Set.discard(x)
        elif order == "check":
            print(1 if x in Set else 0)
        elif order == "toggle":
            if x in Set:
                Set.discard(x)
            else:
                Set.add(x)
```

메모리 문제 때문에 sys까지 사용해주었다.
