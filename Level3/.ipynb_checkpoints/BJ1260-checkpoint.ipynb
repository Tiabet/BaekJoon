DFS와 BFS를 활용하는 기본적인 예제이다. DFS는 재귀문을 통해, BFS는 큐와 반복문을 통해 구현할 수 있다. 이 문제는 두개 다 사용해야 하므로 함수 2개를 따로 선언하여 풀면 될 것 같다.

```python
def DFS(x):
    visited[x]=1
    dfs.append(x)
    for i in node[x]:
        if visited[i]==0:
            DFS(i)

def BFS(x):
    visited[x]=1
    bfs.append(x)
    queue = [x]
    
    while(queue): #queue가 비어있지 않으면 반복실행
        for i in node[queue.pop(0)]: #queue에 들어오자마자 실행되므로 pop해서 꺼내주면 됨
            if(visited[i]==0):
                visited[i]=1
                bfs.append(i)
                queue.append(i) #queue에 append를 해주므로 for문을 다시 실행하게됨

n,m,v = map(int, input().split())

node = [[]for _ in range(n+1)]
visited = [0]*(n+1)
dfs=[]
bfs=[]

for i in range(m):
    a,b=map(int, input().split())
    node[a].append(b)
    node[b].append(a)

for j in range(n+1):
    node[j].sort()

DFS(v)
visited = [0]*(n+1)
BFS(v)

for i in dfs:
    print(i, end=' ')
print()
for j in bfs:
    print(j, end=' ')
```
