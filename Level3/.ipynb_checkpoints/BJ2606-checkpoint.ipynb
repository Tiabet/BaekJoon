     , DFS   .    DFS        .

```python
com = int(input())
net = int(input())

network = [[] for _ in range(com+1)]
count=0
visited = [0] * (com+1)

for _ in range(net):
    n, m = map(int, input().split())
    network[n].append(m)
    network[m].append(n)

def dfs(x):
    visited[x]=1
    for i in network[x]:
        if visited[i]==0:
            visited[i]==1
            dfs(i)

dfs(1)
print(sum(visited)-1)
```
