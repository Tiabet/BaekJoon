이 문제는 절댓값 힙이다. 역시 코드를 약간 수정하면 풀 수 있을 것 같다.

```python
import sys
import heapq

input = sys.stdin.readline

n = int(input())
heap = []

for i in range(n):
    num = int(input())
    if num == 0:
        if heap :
            print(heapq.heappop(heap)[1])
        else :
            print(0)
    else:
         heapq.heappush(heap, (abs(num), num))
```

heap을 2차원 형식으로 구현해서 문제를 풀 수 있었다.
