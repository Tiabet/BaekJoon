이중 우선순위 큐라는 문젠데, 삭제 연산이 두 종류로 나뉜다. 이 점을 유의해서 문제를 풀면 될 것 같다.

```python
from collections import deque

t = int(input())

for _ in range(t):
    dq = deque()
    n = int(input())
    for i in range(n):
        temp = input().split()
        order = temp[0]
        if len(temp) > 1:
            x = int(temp[1])
        
            if order == "I":
                dq.append(x)
            else:
                if x == 1:
                    dq.remove(max(dq))
                else:
                    dq.remove(min(dq))
        else:
            if order == "D" and len(dq) > 0:
                if x == 1:
                    dq.remove(max(dq))
                else:
                    dq.remove(min(dq))
    
    if len(dq) == 0:
        print("EMPTY")
    else:
        print(max(dq), min(dq))
```

```python
import heapq
import sys

input = sys.stdin.readline

t = int(input())

for _ in range(t):
    max_heap = []
    min_heap = []
    visited = [False] * 1000001
    k = int(input())
    
    for i in range(k):
        op, num = input().split()
        num = int(num)
        
        if op == 'I':
            heapq.heappush(max_heap, (-num, i))
            heapq.heappush(min_heap, (num, i))
            visited[i] = True
        elif op == 'D' and num == 1:
            while max_heap and not visited[max_heap[0][1]]:
                heapq.heappop(max_heap)
            if max_heap:
                visited[max_heap[0][1]] = False
                heapq.heappop(max_heap)
        elif op == 'D' and num == -1:
            while min_heap and not visited[min_heap[0][1]]:
                heapq.heappop(min_heap)
            if min_heap:
                visited[min_heap[0][1]] = False
                heapq.heappop(min_heap)
    
    while max_heap and not visited[max_heap[0][1]]:
        heapq.heappop(max_heap)
    while min_heap and not visited[min_heap[0][1]]:
        heapq.heappop(min_heap)
    
    if not max_heap or not min_heap:
        print("EMPTY")
    else:
        print(-max_heap[0][0], min_heap[0][0])
```

여러 시도를 해보았지만 deque 1개로는 시간문제 때문에 문제를 풀 수가 없었다.

```python
from collections import deque

t = int(input())

for _ in range(t):
    dq = deque()
    n = int(input())
    for i in range(n):
        temp = input().split()
        order, x = temp[0], int(temp[1])
        
        if order == "I":
            dq.append(x)
            dq = deque(sorted(dq, reverse=True))
        else:
            if not dq:
                continue
            if x == 1:
                dq.popleft()
            else:
                dq.pop()
    if not dq:
        print("EMPTY")
    else:
        print(dq[0], dq[-1])
```
