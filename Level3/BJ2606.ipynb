이 문제는 읽어보면 깊이 우선 탐색, DFS를 이용해서 푸는 문제다. 몇 번 구현했던 DFS를 살짝만 변형하면 쉽게 풀 수 있을 것 같다.

```python
com = int(input())
net = int(input())

network = [[] for _ in range(com+1)]
count=0
visited = [0] * (com+1)

for _ in range(net):
    n, m = map(int, input().split())
    network[n].append(m)
    network[m].append(n)

def dfs(x):
    visited[x]=1
    for i in network[x]:
        if visited[i]==0:
            visited[i]==1
            dfs(i)

dfs(1)
print(sum(visited)-1)
```
