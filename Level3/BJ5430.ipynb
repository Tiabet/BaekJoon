```python
from collections import deque

n = int(input())

for _ in range(n):
    p = input().strip()
    m = int(input())
    num = input().strip()
    arr = num[1:-1].split(',')
    dq = deque(arr)
    flag = False
    
    if m==0:
        dq=deque([]) #split  ['']      .
        
    R = 0
    
    for i in range(len(p)):
        if p[i]=='R':
            R+=1
        else:
            if len(dq)==0:
                print('error')
                flag = True
                break
            else :
                if R%2==0:
                    dq.popleft()
                else :
                    dq.pop()
    
    if flag :
        continue
    else :
        if R % 2 == 0:
            print('[' + ",".join(dq) + ']')
        else:
            dq.reverse()
            print('[' + ",".join(dq) + ']')
```
