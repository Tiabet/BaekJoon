```python
n = int(input())

stair = []

for i in range(n):
    stair.append(int(input()))

d = [0]*(n+1)

if len(stair)<=2 :
    print(sum(stair))
else :
    d[1]=stair[0]
    d[2]=stair[0]+stair[1]
    for i in range(3,n+1):
        d[i]=max(d[i-3]+stair[i-2]+stair[i-1],d[i-2]+stair[i-1])
    print(d[n])
```
