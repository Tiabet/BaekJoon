이번에도 동적 프로그래밍 문제다. 특이한 점은 시작점은 계단에 포함되지 않는다는 것, 연속으로 밟을 수 있는 최대 계단 수가 2개라는 것이다.

```python
n = int(input())

stair = []

for i in range(n):
    stair.append(int(input()))

d = [0]*(n+1)

if len(stair)<=2 :
    print(sum(stair))
else :
    d[1]=stair[0]
    d[2]=stair[0]+stair[1]
    for i in range(3,n+1):
        d[i]=max(d[i-3]+stair[i-2]+stair[i-1],d[i-2]+stair[i-1])
    print(d[n])
```

동적 프로그래밍의 의미를 살려 d[i]를 채울 때 i-3, i-2를 참고하는 식의 알고리즘이다.
