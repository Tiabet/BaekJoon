경로를 찾는 문제다. 여태 풀었던 문제들과는 유형이 다른 것 같아서 서치를 좀 해보니 플로이드 와샬 알고리즘을 참조하면 되는 문제였다. 최단 경로를 찾는 알고리즘 중엔 다익스트라 알고리즘밖에 모르고 있었는데 찾아보면서 공부를 할 수 있었다.

플로이드 와샬 알고리즘은 한 정점에서 모든 정점으로의 최단거리를 구하는 알고리즘이다. 문제에선 가중치가 없으니 단순하게 방문여부만 체크하면 되어 알고리즘을 응용해 더 쉽게 풀 수 있었다.

```python
n = int(input())

graph = [[]for _ in range(n)]

for i in range(n):
    graph[i]=list(map(int, input().split()))

for k in range(n):
    for i in range(n):
        for j in range(n):
            if graph[i][k]==1 and graph[k][j]==1:
                graph[i][j]=1

for i in range(n):
    for j in range(n):
        print(graph[i][j],end=' ')
    print()
```
