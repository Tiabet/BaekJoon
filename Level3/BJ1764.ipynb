```python
import sys
#input = sys.stdin.readline

listen = []
see = []

n,m = map(int, input().split())

for i in range(n):
    listen.append(input().strip())

for j in range(m):
    see.append(input().strip())
    
intersection = sorted(list(set(listen) & set(see)))

print(len(intersection))
for str in intersection:
    print(str)
```
