문제에도 나와있듯 재귀함수를 이용해서 푸는 문제다. 분할 정복이라는 개념도 쓰인다고 한다. 규칙을 찾아내서 풀 수도 있을 것 같은데, 문제의 취지에 맞게 분할 정복과 재귀함수의 개념을 이용해서 풀어보도록 하겠다.

```python
import sys

#input = sys.stdin.readline()

def dc(x,y,n):
    global count
    if x>r or x+n <= r or y>c or y+n <= c:
        count += n**2
        return 
    
    if n > 2:
        n//=2
        dc(x,y,n)
        dc(x,y+n,n)
        dc(x+n,y,n)
        dc(x+n,y+n,n)
    else:
        if x==r and y==c:
            print(count)
        elif x==r and y+1==c:
            print(count+1)
        elif x+1==r and y==c:
            print(count+2)
        else:
            print(count+3)
        
n,r,c = map(int,input().split())
count = 0
dc(0,0,2**n)
```
