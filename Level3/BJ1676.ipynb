얼핏 보면 이해하기 힘든 문젠데, 팩토리얼 값을 계산한 뒤 0의 개수를 세기만 하면 되는 문제다. 즉 수가 10000 이라면, 답은 4가 된다.

이를 해결하기 위해 10이 몇 번 곱해지는지만 카운트하면 된다. 단 팩토리얼에선 2가 항상 5보다 많기 때문에, 5의 개수를 세면 결국 답에서 0의 개수가 될 것이다.

```python
n = int(input())

zeros = 0

for i in range(2,n+1):
    num = i
    while num>=5:
        if num%5==0:
            zeros+=1
            num=num//5
        else:
            break
        
print(zeros)
```
