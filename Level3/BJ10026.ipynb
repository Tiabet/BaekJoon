그래프 탐색 이론을 사용해 구역을 찾는 문제다. R,G,B 와 RG,B로 나누어서 탐색을 2번 해야하는 것이 특징이다.

```python
n = int(input())

graph = [[] for _ in range(n)]

for i in range(n):
    st = input()
    graph[i] = [char for char in st]

dy = [0,0,1,-1]
dx = [1,-1,0,0]


def dfs(x,y,ch):
    visited[x][y]=1
    for i in range(4):
        nx = x +dx[i]
        ny = y +dy[i]
        if(ny<0 or nx<0 or nx>=n or  ny>=n or visited[nx][ny]==1 or graph[nx][ny]!=ch):
            continue
        dfs(nx,ny,ch)

result = []

for _ in range(2):
    count=0
    visited = [[0]*n for _ in range(n)]
   
    for i in range(0,n):
        for j in range(0,n):
            if(visited[i][j]==0):
                dfs(i,j,graph[i][j])
                count+=1
            if(graph[i][j]=="R"):
                graph[i][j]="G"
    
    result.append(count)
    
for i in result:
    print(i, end=' ')
    
    
```
