자료구조를 사용해서 쉽게 풀 수 있는 문제 같은데, 이런문제의 문제는 항상 메모리다. 메모리 사용량을 줄이기 위해 해쉬를 사용하면 좋다고 한다. 파이썬의 dict 함수가 이미 hashmap의 기능을 사용하고 있으므로 이를 사용하면 쉽게 풀 수 있을 문제다.

```python
n, m = map(int, input().split())

pocketmon = {}

def get_key(val):
    for key, value in pocketmon.items():
        if val == value:
            return key

for i in range(n):
    pocketmon[input()] = i+1
    
result = []
    
for j in range(m):
    order = input()
    if order.isdigit():
        print(get_key(order))
    else:
        print(pocketmon[order])
```
