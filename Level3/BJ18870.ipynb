좌표를 압축한다는 문젠데, 대소를 비교해서 자신보다 작은 수가 얼마나 있는지 카운트하는 문제다.

```python
n = int(input())

lst = list(map(int, input().split()))

result = []

for i in range(n):
    count = 0
    for j in range(n):
        if lst[i]>lst[j]:
            count+=1
    result.append(count)

print(*result)
```

이런식으로 이중포문으로 코드를 짜니 시간초과가 발생했다. 그래서 다른 방법을 생각해보았다.

```python
n = int(input())
coords = list(map(int, input().split()))

# 각 좌표의 등장 횟수를 센다
freq = {}
for coord in coords:
    if coord in freq:
        continue
    else:
        freq[coord] = 1

# 등장 횟수를 누적하여 압축 결과를 계산한다
compressed_coords = {}
acc = 0
for coord in sorted(freq.keys()):
    compressed_coords[coord] = acc
    acc += freq[coord]

# 각 좌표의 압축 결과를 출력한다
print(' '.join(str(compressed_coords[coord]) for coord in coords))
```

같은 좌표에 대해 셀 필요가 없으니 dict를 이용해서 풀 수 있었다.
