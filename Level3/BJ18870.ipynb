```python
n = int(input())

lst = list(map(int, input().split()))

result = []

for i in range(n):
    count = 0
    for j in range(n):
        if lst[i]>lst[j]:
            count+=1
    result.append(count)

print(*result)
```

```python
n = int(input())
coords = list(map(int, input().split()))

#     
freq = {}
for coord in coords:
    if coord in freq:
        continue
    else:
        freq[coord] = 1

#      
compressed_coords = {}
acc = 0
for coord in sorted(freq.keys()):
    compressed_coords[coord] = acc
    acc += freq[coord]

#     
print(' '.join(str(compressed_coords[coord]) for coord in coords))
```
