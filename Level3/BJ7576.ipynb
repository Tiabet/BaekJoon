토마토가 익어가는 것을 확인하는 문제인데 탐색 문제의 응용으로 보인다.BFS를 사용해서 문제를 풀면 될 것 같다.

```python
from collections import deque

m, n = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)] #매트릭스 생성
queue = deque([]) #빈 deque를 선언
dx, dy = [-1, 1, 0, 0], [0, 0, -1, 1] #좌우하상
res = 0 #흐른 시간

for i in range(n):
    for j in range(m):
        if matrix[i][j] == 1:
            queue.append([i, j]) #토마토가 있는 곳 큐에 저장

def bfs():
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx, ny = dx[i] + x, dy[i] + y
            if 0 <= nx < n and 0 <= ny < m and matrix[nx][ny] == 0:
                matrix[nx][ny] = matrix[x][y] + 1 #1이면 2를, 2면 3을 0 위치에 저장
                queue.append([nx, ny])#다음 탐색할 곳을 큐에 추가

bfs()
for i in matrix:
    for j in i:
        if j == 0:
            print(-1)
            exit(0)
    res = max(res, max(i))
print(res - 1)
```

이전엔 deque 없이 그냥 pop(0)을 이용해 풀었는데, 시간 초과 문제 때문에 deque를 사용해야 했다. deque는 스택과 큐의 기능을 동시에 갖고 있는 함수이다. 데이터구조론 시간에 배운 적이 있었다.  (double-ended-queue) 그리고 상하좌우로 움직이는 것을 표현하기 위해 dx, dy를 선언하고, res는 흐른 시간을 저장하기 위해 선언했다. exit(0) 함수를 사용했는데 이는 코드를 종료할 때 사용하는 코드다.
