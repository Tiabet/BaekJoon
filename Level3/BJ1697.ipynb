최단경로를 찾는 알고리즘인데, +1, -1, x2 로 할 수 있는 선택지가 많아서 살짝 까다로워보인다. 그래프 이론이기 때문에 너비 우선 탐색 같은 알고리즘을 사용해야 한다고 하는데, 적용하는 방법을 좀 고민해보았다.

```python
from collections import deque

def bfs(n, k):
    queue = deque()
    queue.append((n, 0)) 
    visited = [False] * 100001  

    while queue:
        x, time = queue.popleft()
        visited[x] = True

        if x == k:  
            return time

        
        if x - 1 >= 0 and not visited[x-1]:
            queue.append((x-1, time+1))
        if x + 1 <= 100000 and not visited[x+1]:
            queue.append((x+1, time+1))

        
        if 2*x <= 100000 and not visited[2*x]:
            queue.append((2*x, time+1))

n, k = map(int, input().split())
print(bfs(n, k))
```

bfs 의 특징을 이용해 가장 빨리 도착한 경로의 시간을 반환하는 문제다. queue에 있는 모든 경로를 다 탐색하기 때문에 이런 식으로 효과적으로 풀 수 있다.
