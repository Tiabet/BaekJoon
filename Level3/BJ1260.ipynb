```python
def DFS(x):
    visited[x]=1
    dfs.append(x)
    for i in node[x]:
        if visited[i]==0:
            DFS(i)

def BFS(x):
    visited[x]=1
    bfs.append(x)
    queue = [x]
    
    while(queue): #queue   
        for i in node[queue.pop(0)]: #queue   pop  
            if(visited[i]==0):
                visited[i]=1
                bfs.append(i)
                queue.append(i) #queue append  for  

n,m,v = map(int, input().split())

node = [[]for _ in range(n+1)]
visited = [0]*(n+1)
dfs=[]
bfs=[]

for i in range(m):
    a,b=map(int, input().split())
    node[a].append(b)
    node[b].append(a)

for j in range(n+1):
    node[j].sort()

DFS(v)
visited = [0]*(n+1)
BFS(v)

for i in dfs:
    print(i, end=' ')
print()
for j in bfs:
    print(j, end=' ')
```
