좌표를 정렬하는 문제인데, 핵심은 x좌표를 우선으로 정렬하되 동수가 있으면 y좌표를 기준으로 정렬한다는 것이다.

```python
n = int(input())
result = []

for _ in range(n):
    x,y = map(int,input().split())
    result.append([x,y])
    
result.sort(key=lambda x : (x[0],x[1]))

for i in range(n):
    print(result[i][0],result[i][1])
```

파이썬의 sort 함수를 통해 쉽게 정렬을 할 수 있었다. key = lambda 로 하고 x : (x[0],x[1]) 처럼 코딩을 하면, x[0]을 우선으로 고려하고 후순위로  x[1]을 고려함을 알 수 있다. 오름차순이 아닌 내림차순으로 하고 싶으면 앞에 - 부호를 붙이면 된다.
