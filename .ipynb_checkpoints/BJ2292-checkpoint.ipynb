이 문제의 핵심은 숫자가 1을 기준으로 몇 번째 칸에 있느냐 하는 것인데, 1을 중심으로 칸의 수가 1, 6, 12, 18, 24 ... 로
6의 배수만큼 그 수가 증가함을 알 수 있다. 따라서 찾고자 하는 수가 몇 번째에 있는 수인지만 체크하면 된다.

```python
n = int(input())
reduce = 6
count = 1

if n==1:
   print(count) 
else :
    n-=1
    while n>0:
        n-=reduce
        reduce+=6
        count+=1
    print(count)
    
```

빼는 수를 6씩 늘려가면서 가볍게 풀 수 있었다.
